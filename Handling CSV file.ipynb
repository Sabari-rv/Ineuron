```python
import csv
csv_file = "bitcoin.csv"
print("File name: ",csv_file)
fields = []  #initializing fields
rows = [] #initializing rows
with open(csv_file,'r') as csvfile:
    csvreader = csv.reader(csvfile) #creating csv reader object
    fields = next(csvreader) #extracting fieldname through first row
    for row in csvreader: #extracting each data row one by one
        rows.append(row)
    print("Fields: ", fields) #printing fields
print("Number of rows in file:", csvreader.line_num) #printing number of rows
print("Field name: " + ",".join(field for field in fields)) #printing column names
print("Parsing each column of row: ")
for row in rows[:5]:
    for col in row:
        print("%s"%col,end=" ")
    print("\n")
```
